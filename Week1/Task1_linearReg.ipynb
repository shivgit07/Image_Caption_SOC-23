```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# Loading the iris dataset
iris = load_iris()
x = iris.data
y = iris.target

partition_percentages = range(10, 90, 2)

losses = []
train_size = []


for percentage in partition_percentages:
    #splitting
    x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=percentage/100, random_state=42)
    
    #linear regression model and fitting
    model = LinearRegression()
    model.fit(x_train, y_train)
    
    #predicted y
    y_pred = model.predict(x_test)
    
    #mean squared error loss
    loss = mean_squared_error(y_test, y_pred)
    
    losses.append(loss)
    train_size.append(len(x_train))

    perc_part = []
    for ele in train_size: 
      perc_part.append(((ele)/(len(x)))*100)
```

```python
#plotting loss vs % partitioned
plt.plot(perc_part, losses)
plt.xlabel('% partitioned')
plt.ylabel('Loss')
plt.title('Loss vs % partitioned')
plt.show()
```

```python
# Finding the minimum loss and corresponding training size
min_loss = min(losses)
min_loss_index = losses.index(min_loss)
optimal_training_size = train_size[min_loss_index]
optimal_partition_percentage = partition_percentages[min_loss_index]

print(f"Minimum loss: {min_loss}")
print(f"Optimal training size: {optimal_training_size}")
print(f"Optimal partition percentage: {optimal_partition_percentage}")
```
